```python
# Exploring the data
dataset=pd.read_csv("SMSSpamCollection.tsv",sep='\t',header=None)
dataset.columns = ['label', 'body_text']
dataset.head()
```

```python
#How many rows and colums in the dataset 
print("input datya has {} rows and {} column ".format(len(dataset), len(dataset.columns)))
```

```python
# How many Spam and HAM in dataset
print("out od {} rows , {} are Spam and {} are ham".format(len(dataset), 
                                                           len(dataset[dataset['label']=='spam']),
                                                          len(dataset[dataset['label']=='ham'])))
```
